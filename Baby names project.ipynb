```python
# Finding out what the most popular baby names are in the US
```

```python
# import libraries and read csv file

import pandas as pd
import numpy as np

file = pd.read_csv('babyNamesUSYOB-full.csv')
file
```

```python
file['YearOfBirth'].value_counts()
```

```python
file['Name'].value_counts()
```

```python
file['Sex'].value_counts()
```

```python
# clean data by having all names as lower or uppercase 
file['Name'] = file['Name'].str.upper()
file.head()
```

```python
file = file.drop(['YearOfBirth'] , axis = 1)
```

```python
# remove all duplicate names
file.drop_duplicates(keep = False,  inplace = True)
file
```

```python
file.sort_values('Number', ascending = False)
```

```python
import matplotlib.pyplot as plt
from wordcloud import WordCloud

All_babies = dict(zip(file['Name'], file['Name'].value_counts()))
```

```python
wordcloud =  WordCloud(background_color = "white")
wordcloud.generate_from_frequencies(frequencies = All_babies)
plt.figure()
plt.imshow(wordcloud)
plt.show()
```
