```python
# given a dataset that shows which users have purchased an iphone, 
# predict if the customer will
# purchase an iphone or not given their gender, age and salary
```

```python
import pandas as pd
import numpy as np
import requests
import io
```

```python
url = 'https://raw.githubusercontent.com/omairaasim/machine_learning/master/project_11_k_nearest_neighbor/iphone_purchase_records.csv'
download = requests.get(url).content
df = pd.read_csv(io.StringIO(download.decode('utf-8')))
df.head()
```

```python
df.info()
```

```python
# convert gender to numerical values using label encoder
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df['Gender']= le.fit_transform(df['Gender'])
df


```

```python
# Let x. be gender, age and salary
# let y be purchase iphone

x = df.iloc[:, 0:3].values
y = df.iloc[:, -1].values

```

```python
# split data into training and testing set
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.25, random_state = 0)

print(x_train)
```

```python
# feature scaling
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train =sc.fit_transform(x_train)
x_test = sc.transform(x_test)


```

```python
# fit KNN Classifier
from sklearn.neighbors import KNeighborsClassifier
# metric = minkowski and p=2 is Euclidean Distance
# metric = minkowski and p=1 is Manhattan Distance
classifier = KNeighborsClassifier(n_neighbors = 5, metric="minkowski", p=2)
classifier.fit(x_train, y_train)
```

```python
y_pred = classifier.predict(x_test)
```

```python
# confusion metrics
from sklearn import metrics
cm = metrics.confusion_matrix(y_test, y_pred)
print (cm)
```

```python
accuracy = metrics.accuracy_score(y_test, y_pred)
print("Accuracy score:",accuracy)
```

```python
precision = metrics.precision_score(y_test, y_pred) 
print("Precision score:",precision)
```

```python
recall = metrics.recall_score(y_test, y_pred) 
print("Recall score:",recall)
```
