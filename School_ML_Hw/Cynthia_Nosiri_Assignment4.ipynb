```python
# task breakdown
# display rows with more than 15% missing data
#     -----first get total no. of missing data in a row
#         -----then total no. data in a row or total no. of columns
#     ------% of missing data in each row
#     -------filter for less or more than 15
```

```python
# display rows with more than 15% missing data
for i in range(len(df.index)):
    percent = df.iloc[i].isnull().sum()/len(df.columns) * 100
    if percent > 15:
        print("More than 15% is missing in row ", i ,":", percent.round(0), "%")
    if percent < 15 and percent > 0:
        print("Less than 15% is missing in row ", i ,":", percent.round(0), "%")
```

```python
# drop all rows with more than 15% missing data and display no. of rows dropped
# data i'll like to keep(rows with 15% or less missing data)
new_df= df.loc[df.isna().mean(axis = 1) < 0.15]
new_df.shape[0]
```

```python
print(df.shape[0] - new_df.shape[0], "observations dropped due to having more than 15% missing feature points")
```

```python
# creating a dataframe new_df1 with only features
new_df1 = new_df.drop(columns = [49])
new_df1
```

```python
# For rows with less than 15% data, replace missing data with the median value of that feature and 
# display the row/column position of cell with missing feature and the median value inserted

# steps: loop through the rows and columns, if there are missing values in the rows and columns, 
# print that position with the median

for i in new_df1.index:
    for j in new_df1.columns:
        if new_df1.isna().loc[i][j] == True:
            print("Replace nan at position", (i, j), "with {} ".format(new_df1.median(numeric_only = True)[j]))
    
```

```python
# replace the null values with their corresponding median
new_df.fillna(new_df.median(numeric_only = True), inplace = True)
new_df
```

```python
# Build and test classifiers 
from sklearn.preprocessing import MinMaxScaler

# new_df3 contains only features
# target contains only the target column
new_df3 = new_df2.drop(columns = 49)
target = new_df2[49]
```

```python
scaler = MinMaxScaler()
scaled_data = scaler.fit_transform(new_df3)

# put scaled new_df3 back into a dataframe
new_df3 = pd.DataFrame(scaled_data)
new_df3
```

```python
# add the target to new_df3
# target.tolist()
# new_df3[49] = target
# new_df3
```

```python
X  = new_df3
Y = target
```

```python
# train/test split data
from sklearn.model_selection import train_test_split
X_train, x_test, Y_train, y_test = train_test_split(X, Y, test_size = 0.3, random_state = 109)
```

```python
# multinomialNB Classifier Model

from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score
mnb = MultinomialNB()
mnb.fit(X_train, Y_train)

predicted = mnb.predict(x_test)

score = accuracy_score(y_test, predicted)
print(score)
```

```python
# SVM classifier model
from sklearn.svm import SVC
svm = SVC(kernel = 'linear', C= 1)
svm.fit(X_train, Y_train)
y_pred = svm.predict(x_test)
accuracy_score(y_test, y_pred)
```
